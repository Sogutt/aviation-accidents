```python
# Create a dataframe that has each US state and their corresponding abbreviations 
df3 = pd.DataFrame.from_dict(us_state_abbrev, orient='index')
df3
```

```python
# Join the abreviation dataframe and the census population data
df2_abrv = df2.join(df3, on='Geography')
df2_abrv
```

```python
# Now have a df containing the state abreviations and the population estimates
state_abrv = df2_abrv[0]
df2_abrv['state_abrv'] = state_abrv

population = df2_abrv['Population Estimate - July 1, 2017']
df2_abrv['population_estimate'] = population
df2_abrv
```

```python
# drop the irrelevant 2010 estimate
df2_abrv.drop(['Id2', 'Population Estimate - April 1, 2010 Estimates Base', 'Population Estimate - July 1, 2017',0], axis=1, inplace=True)
#df2_abrv.drop(['Geography'], axis=1, inplace=True)
df2_abrv
```

```python
# We now have a final dataframe that can be used for the state normalization
df_final = df.merge(df2_abrv, left_on='Event State',  right_on='state_abrv')
df_final
```
